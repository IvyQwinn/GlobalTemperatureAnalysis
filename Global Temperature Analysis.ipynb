```python
import pandas as pd

# Load datasets
global_temp = pd.read_csv('GLB.Ts+dSST.csv')
northern_hemisphere_temp = pd.read_csv('NH.Ts+dSST.csv')
southern_hemisphere_temp = pd.read_csv('SH.Ts+dSST.csv')
zonal_temp = pd.read_csv('ZonAnn.Ts+dSST.csv')

# Display the first few rows of each dataset
(global_temp.head(), northern_hemisphere_temp.head(), southern_hemisphere_temp.head(), zonal_temp.head())
```

```python
import pandas as pd

# Check structure and summary of each dataset

# Print structure 
print("Structure of global_temp:")
print(global_temp.info())  # This will show column names, data types, and non-null counts

print("Structure of northern_hemisphere_temp:")
print(northern_hemisphere_temp.info())

print("Structure of southern_hemisphere_temp:")
print(southern_hemisphere_temp.info())

print("Structure of zonal_annual_means:")
print(zonal_temp.info())

# Print summary statistics 
print("Summary of global_temp:")
print(global_temp.describe())  # This will show summary statistics like mean, min, max, etc.

print("Summary of northern_hemisphere_temp:")
print(northern_hemisphere_temp.describe())

print("Summary of southern_hemisphere_temp:")
print(southern_hemisphere_temp.describe())

print("Summary of zonal_annual_means:")
print(zonal_temp.describe())
```

```python
# Handle missing values 
global_temp.dropna(inplace=True)
northern_hemisphere_temp.dropna(inplace=True)
southern_hemisphere_temp.dropna(inplace=True)
zonal_temp.dropna(inplace=True)

# Convert Year column to numeric
global_temp['Year'] = pd.to_numeric(global_temp['Year'], errors='coerce')
northern_hemisphere_temp['Year'] = pd.to_numeric(northern_hemisphere_temp['Year'], errors='coerce')
southern_hemisphere_temp['Year'] = pd.to_numeric(southern_hemisphere_temp['Year'], errors='coerce')
zonal_temp['Year'] = pd.to_numeric(zonal_temp['Year'], errors='coerce')
```

```python
import pandas as pd

global_temp_cleaned = global_temp.copy()

# Convert monthly columns to numeric, coerce errors to NaN
monthly_columns = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
global_temp_cleaned[monthly_columns] = global_temp_cleaned[monthly_columns].apply(pd.to_numeric, errors='coerce')

# Calculate the Annual Mean
global_temp_cleaned['Annual_Mean'] = global_temp_cleaned[monthly_columns].mean(axis=1, skipna=True)

```

```python
import pandas as pd

northern_hemisphere_temp_cleaned = northern_hemisphere_temp.copy()

# Convert monthly columns to numeric, coerce errors to NaN
monthly_columns = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
northern_hemisphere_temp_cleaned[monthly_columns] = northern_hemisphere_temp_cleaned[monthly_columns].apply(pd.to_numeric, errors='coerce')

# Calculate the Annual Mean
northern_hemisphere_temp_cleaned['Annual_Mean'] = northern_hemisphere_temp_cleaned[monthly_columns].mean(axis=1, skipna=True)
```

```python
import pandas as pd

southern_hemisphere_temp_cleaned = southern_hemisphere_temp.copy()

# Convert monthly columns to numeric, coerce errors to NaN
monthly_columns = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
southern_hemisphere_temp_cleaned[monthly_columns] = southern_hemisphere_temp_cleaned[monthly_columns].apply(pd.to_numeric, errors='coerce')

# Calculate the Annual Mean
southern_hemisphere_temp_cleaned['Annual_Mean'] = southern_hemisphere_temp_cleaned[monthly_columns].mean(axis=1, skipna=True)


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example of plotting global temperature trends
plt.figure(figsize=(10, 6))
sns.lineplot(data=global_temp_cleaned, x='Year', y='Annual_Mean')
plt.title('Global Annual Mean Temperature Anomalies')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (C)')
plt.grid(True)
plt.show()

# Example of plotting Northern Hemisphere temperature trends
plt.figure(figsize=(10, 6))
sns.lineplot(data=northern_hemisphere_temp_cleaned, x='Year', y='Annual_Mean')
plt.title('Northern Hemisphere Annual Mean Temperature Anomalies')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (C)')
plt.grid(True)
plt.show()

# Example of plotting Southern Hemisphere temperature trends
plt.figure(figsize=(10, 6))
sns.lineplot(data=southern_hemisphere_temp_cleaned, x='Year', y='Annual_Mean')
plt.title('Southern Hemisphere Annual Mean Temperature Anomalies')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (C)')
plt.grid(True)
plt.show()

```
